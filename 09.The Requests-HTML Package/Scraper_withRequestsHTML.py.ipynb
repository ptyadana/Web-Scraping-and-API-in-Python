# Set up


```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
```

```python
response = session.get('https://en.wikipedia.org/wiki/Association_football')
response.status_code
```

```python
response.html
```

### Links

```python
urls = response.html.links
urls
```

#### getting links with absolute path

```python
full_paths_url = response.html.absolute_links
full_paths_url
```

```python
type(urls)
```

### Searching for elements

```python
links = response.html.find('a')
links
```

```python
links[5]
```

```python
# to get raw html
links[5].html
```

```python
links[5].attrs
```

#### to get links with keyword

```python
response.html.find('a', containing='wikipedia')
```

```python
[tag.text for tag in response.html.find('a', containing='wikipedia')]
```

#### finding first element

```python
response.html.find('p', first = True)
```
