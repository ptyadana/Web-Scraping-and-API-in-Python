# Set up


```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
```

```python
r = session.get('https://en.wikipedia.org/wiki/Association_football')
r.status_code
```

```python
r.html
```

## CSS Selectors

### Select elements based on tag name

```python
r.html.find('span')
```

### Select elements based on ID

```python
r.html.find('#Name')
```

```python
#case sensitive
r.html.find('#name')
```

```python
# as Id is specific to one element, we can set first=True as we are sure that there is only one item in results

r.html.find('#Duration_and_tie-breaking_methods', first=True)
```

## Selecting by class

```python
r.html.find('.mw-headline')
```

```python
r.html.find('.metadata')
```

### Searching elements with multiple classes

```python
r.html.find('.metadata.plainlinks')
```

----------

## Selecting based on other attributes

```python
r.html.find('[target]')
```

```python
r.html.find('[role=note]')
```

### Searching href where it contains keyword 'wikipedia'

```python
r.html.find('[href*=wikipedia]')
```

----------

### Combining different filters together

```python
r.html.find('a[href*=wikipedia]')
```

### Searching for  tag with class 'internal'

```python
r.html.find('a.internal')
```

### Searching for div with class names 'thumb' and 'tright'

```python
r.html.find('div.thumb.tright')
```

### div with role=note and classes with 'hatnote', 'navigation-not-searchable'

```python
r.html.find("div[role=note][class='hatnote navigation-not-searchable']")
```

--------

## Incorporating tag hierarchy

### every span insides h2

```python
r.html.find('h2 span')
```

### direct child h2 of div

```python
r.html.find('div > h2')
```

## Close session

```python
session.close()
```
